```python
# Custom function to modify range of SST for HYCOM
def scaleAndOffset_HYCOM_SST(image):
    return ee.Image(image).multiply(0.001).add(20.0)
    
```

```python
# HYCOM SST collection
collection = (ee.ImageCollection('HYCOM/sea_temp_salinity')
    .filter(ee.Filter.date("2019-03-05","2019-03-09"))
    .select('water_temp_0')
)

min_val = (4-20)*1000
max_val = (27-20)*1000
vis_params = {
  "min": min_val,
  "max": max_val,
  "palette": ['000000', '005aff', '43c8c8', 'fff700', 'ff0000'],
}

# Add slider (2/day)
m.add_time_slider(collection, vis_params, time_interval=1)
```

```python
# Scaled
collection_scaled = collection.map(scaleAndOffset_HYCOM_SST)

vis_params_scaled = {
  "min": 4.0,
  "max": 25.0,
  "palette": ['000000', '005aff', '43c8c8', 'fff700', 'ff0000'],
}

# Add slider (2/day), which currently gives an error: TraitError: setting min > max
#m.add_time_slider(collection_scaled, vis_params_scaled, time_interval=1)
```

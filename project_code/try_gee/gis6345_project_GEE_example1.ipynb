```python
# Imports
import ee
import geemap
```

```python
ee.Authenticate()
```

```python
m = geemap.Map()
```

```python
m

```

```python
# ROI
roi = m.user_roi
if roi is None:
    #roi = ee.Geometry.BBox(-76.0, 34.25, -73.0, 36.5)
    roi = ee.Geometry.BBox(-77.0, 30.0, -60.0, 45.0)
    m.add_layer(roi)
    m.center_object(roi)
    
```

```python
timelapse = geemap.modis_ocean_color_timelapse(
    satellite="Aqua",
    start_date="2017-01-01",
    end_date="2020-12-31",
    roi=roi,
    out_gif="modis.gif",
    frequency="month",
)
geemap.show_image(timelapse)
```
